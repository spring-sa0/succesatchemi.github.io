```python
# wifi_password_gen.py
import secrets
import string
import argparse

def generate_password(length=16, use_symbols=True):
    if length < 8 or length > 63:
        raise ValueError("La longueur doit être entre 8 et 63.")
    alphabet = string.ascii_letters + string.digits
    if use_symbols:
        alphabet += "!@#$%^&*()-_=+[]{}<>?/"
    # Garantir au moins une majuscule, une minuscule, un chiffre
    while True:
        pwd = ''.join(secrets.choice(alphabet) for _ in range(length))
        if (any(c.islower() for c in pwd) and
            any(c.isupper() for c in pwd) and
            any(c.isdigit() for c in pwd)):
            return pwd

if __name__ == "__main__":
    p = argparse.ArgumentParser(description="Génère un mot de passe Wi-Fi sécurisé")
    p.add_argument("-l", "--length", type=int, default=16, help="longueur (8-63)")
    p.add_argument("--no-symbols", action="store_true", help="ne pas inclure de symboles")
    args = p.parse_args()
    print(generate_password(args.length, not args.no_symbols))
```
